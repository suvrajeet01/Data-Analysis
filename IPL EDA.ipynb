```python
def batsman_vs_teams(batsman_name):
    mask=delivery['batsman']==batsman_name
    vk=delivery[mask]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def batsman_vs_bowler(batsman_name,bowler_name):
    mask1=delivery['batsman']==batsman_name
    mask2=delivery['bowler']==bowler_name
    vk1=delivery[mask1]
    vk2=delivery[mask2]
    return vk1.groupby('vk2')['batsman_runs'].sum().sort_values(ascending=False)
```
